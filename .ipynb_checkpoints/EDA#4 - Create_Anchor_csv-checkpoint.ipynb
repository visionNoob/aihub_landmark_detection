```python
import sys 
import os 
sys.path.append(os.pardir)
import os.path as osp 
import math 

import pandas as pd 
from glob import glob 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from tqdm import tqdm 
import yaml 
from PIL import Image 
```

```python
trainDir_list = ['train', 'train2', 'test', 'test2', 'val', 'val2']
```

```python
motherDir = osp.abspath('../..')
dataDir = osp.join(motherDir, 'data', 'refined', trainDir_list[0])
dataList = os.listdir(dataDir)
```

```python
for path in tqdm(dataList):
    target = osp.join(dataDir, path)
    yamls = glob(osp.join(target, '*.yaml'))
    imgs = glob(osp.join(target, '*.jpg'))


    if len(yamls) != len(imgs):
        print("mismathced")
    
    else: 
        pass 

print("Well-refined")
```

*** 
### check yamls 

```python
landmark = [] 
xmin, ymin, xmax, ymax = [], [], [], []
fileName = [] 

for path in tqdm(dataList):
    target = osp.join(dataDir, path)
    yamls = glob(osp.join(target, '*.yaml'))
    
    for item in yamls: 
        item_yaml = item 
        
        with open(item_yaml) as file: 
            ds = yaml.load(file)
            
 
        
        fileName.append(item_yaml.split("/")[-1].split(".")[0])
        landmark.append(ds['regions'][0]['class'])
            
        x1, y1, x2, y2 = ds['bbox']['x1'], ds['bbox']['y1'], ds['bbox']['x2'], ds['bbox']['y2']
        xmin.append(x1)
        ymin.append(y1)
        xmax.append(x2)
        ymax.append(y2)
        
df = pd.DataFrame({ 'filename'  : fileName,
                    'width' : 640,
                    'height': 480,
                    'class' : landmark,
                    'xmin' : xmin,
                    'ymin' : ymin,
                    'xmax' : xmax,
                    'ymax' : ymax})

df.to_csv('Anchor.csv', sep=',', na_rep='NaN')
```
